```python
info = pd.read_csv(path+'Info.txt', sep=' ')
info = info.drop('Unnamed: 7', axis=1)
info
```

```python
info.dropna(subset=['SEVERITY'], inplace=True)
info.reset_index(inplace=True)
info
```

```python
info = info.drop([3], axis=0)
info.reset_index(inplace=True)
info
```

```python
ids = {}
for i in range(len(info)):
    ids[i] = info.REFNUM[i]
```

```python
ids
```

```python
label = []
for i in range(len(info)):
    if info.SEVERITY[i] == 'B':
        label.append(1)
    else:
        label.append(0)
```

```python
label = np.array(label)
```

```python
label
```

```python
img_name = []
for i in range(len(label)):
    img_name.append(path+info.REFNUM[i]+'.pgm')
```

```python
img_name = np.array(img_name)
```

```python
img_name
```

```python
import matplotlib.pyplot as plt
import cv2

img_path = []
last_label = []

for i in range(len(img_name)):
    img = cv2.imread(img_name[i], 0)
    img = cv2.resize(img, (90, 90))
    rows, cols = img.shape
    for angle in range(360):
        M = cv2.getRotationMatrix2D((cols/2, rows/2), angle, 1)
        img_rotated = cv2.warpAffine(img, M, (90, 90))
        img_path.append(img_rotated)
        if label[i]==1:
            last_label.append(1)
        else:
            last_label.append(0)
```

```python
x_train, x_test, y_train, y_test = train_test_split(img_path, last_label, test_size = 0.2, random_state = 42)
```

```python
nsample, nx, ny = x_train.shape
x_train_2d = x_train.reshape((nsample, nx*ny))
x_train_2d.ndim
```

```python
nsample2, mx, my = x_test.shape
x_test_2d = x_test.reshape((nsample2, mx*my))
x_test_2d.ndim
```

```python
from sklearn import svm
clf = svm.SVC(kernel='rbf')
clf.fit(x_train_2d, y_train)
y_pred = clf.predict(x_test_2d)
```
